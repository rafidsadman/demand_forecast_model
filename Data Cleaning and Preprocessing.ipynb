### Preparing the Weather Data and merging with *by_date* dataframe

```python
#reading the weather data
weather = pd.read_csv("weather_data_london.csv")
```

```python
df2 = weather[["Year", "Month", "Date"]].copy()
df2.columns = ["year", "month", "day"]

dates = pd.to_datetime(df2)
weather["date"] = dates
```

```python
#dropping the unnecessary variables
weather.drop(["Year", "Month", "Date"], 1, inplace=True)
```

```python
weather.head()
```

```python
# Merging of the datasets
merged = pd.merge(weather, by_date, how = "inner", on = "date")
```

```python
merged.head()
```

### Adding Extra Explanatory Variables

```python
#getting the weekday names
x = []
for j in merged["date"]:
    x.append(j.strftime("%A"))

merged["weekday"] = x
```

#### *Adding holidays and weekends in the dataset* 

```python
#adding the holidays
uk_holidays = holidays.UK()

holi = []
for i in merged["date"]:
    holi.append(i in uk_holidays)

merged["holiday_today"] = holi
```

```python
#adding the weekends
weekends = []
for i in merged["weekday"]:
    if (i == "Saturday"):
        weekends.append(True)
    elif(i == "Sunday"):
        weekends.append(True)
    else:
        weekends.append(False)

merged["weekend"] = weekends
```

```python
#Setting up the conditions and the values for "offday"
conditions = [
    (merged["weekend"] == True) | (merged["holiday_today"] == True),
    (merged["weekend"] == False) & (merged["holiday_today"] == False),
    ]

values = [True, False]
#adding offday(holiday or weekend) in the dataset
merged["offday"] = np.select(conditions, values)
```

#### *Adding a column that tells whether the next day is an offday(holiday or weekend)*

```python
indices = list(range(0,(len(merged)-1)))
next_day_holiday = []
for i in indices:
    if (merged["offday"][i+1] == 1):
        next_day_holiday.append(True)
    else:
        next_day_holiday.append(False)

next_day_holiday.append(False)

merged["offday_tomorrow"] = next_day_holiday
```

```python
#adding month as a categorical variable
merged["month"] = merged["date"].dt.month
merged["month"] = merged["month"].astype("category")
```

```python
#Finding the timedelta - days since first operation
merged["delta_days"] = merged["date"] - merged["date"][0]
merged["delta_days"] = merged["delta_days"].dt.days
```

```python
#finding the number of months since inception
floors_months = []
for i in merged["delta_days"]:
    floors_months.append(math.floor(i/30))
    
merged["delta_months"] = floors_months
```

```python
#finding the number of weeks since inception
floors_weeks = []
for i in merged["delta_days"]:
    floors_weeks.append(math.floor(i/7))
    
merged["delta_weeks"] = floors_weeks
```

```python
#finding the number of years since inception
floors_years = []
for i in merged["delta_days"]:
    floors_years.append(math.floor(i/365))
    
merged["delta_years"] = floors_years
```

```python
### Dropping unnecessary variables
drop_vars = ["holiday_today"]
merged.drop(drop_vars, 1, inplace = True)
```

### Changing the format of the variables to appropriate forms

```python
merged["weekday"] = merged["weekday"].astype("category")
merged["weekend"] = merged["weekend"].astype("category")
merged["offday"]  = merged["offday"].astype("category")
merged["offday_tomorrow"] = merged["offday_tomorrow"].astype("category")
```

```python
#taking "date" to left most column
dates = merged["date"]
merged.drop("date", 1, inplace = True)
merged.insert(0, "date", dates)
```

```python
#taking "revenue" to right-most column
revenues = merged["revenue"]
merged.drop("revenue", 1, inplace = True)
merged.insert(14, "revenue", revenues)
```
